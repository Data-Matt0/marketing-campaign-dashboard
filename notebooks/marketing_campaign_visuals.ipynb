```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
url = 'https://raw.githubusercontent.com/Data-Matt0/marketing-campaign-dashboard/refs/heads/main/data/queries/q1_response_by_education.csv'
df = pd.read_csv(url)

# Set plot style
sns.set(style='whitegrid')

# Set figure size and color palette
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=df,
    x='education',
    y='response_rate_percent',
    palette='viridis'
)

# Add value labels on top of each bar
for index, row in df.iterrows():
    barplot.text(index, row.response_rate_percent + 0.5, f"{row.response_rate_percent:.1f}%",
                 color='black', ha='center', fontsize=10)

# Titles and labels
plt.title('📊 Response Rate by Education Level', fontsize=16, weight='bold')
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Response Rate (%)', fontsize=12)

# Improve layout
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
