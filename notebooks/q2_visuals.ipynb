```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url = 'https://raw.githubusercontent.com/Data-Matt0/marketing-campaign-dashboard/refs/heads/main/data/queries/q2_response_by_income.csv'
df = pd.read_csv(url)

# Set Seaborn style
sns.set(style='whitegrid', font_scale=1.1)

# Create figure
plt.figure(figsize=(12, 7))
barplot = sns.barplot(
    data=df,
    x='income_bracket',
    y='response_rate_percent',
    palette='rocket'
)

# Annotate each bar correctly using bar container positions
for bar, value in zip(barplot.patches, df['response_rate_percent']):
    x = bar.get_x() + bar.get_width() / 2
    y = bar.get_height()
    barplot.text(x, y + 0.5, f"{value:.1f}%", ha='center', color='black', fontsize=11, weight='bold')

# Titles and labels
plt.title('💰 Response Rate by Income Bracket', fontsize=18, weight='bold')
plt.xlabel('Income Bracket', fontsize=13, weight='bold')
plt.ylabel('Response Rate (%)', fontsize=13, weight='bold')

# Axis adjustments
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.tight_layout()

# Save and show
plt.savefig('response_rate_by_income.png', dpi=300, bbox_inches='tight')
plt.show()
```
