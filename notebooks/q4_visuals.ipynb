```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV from GitHub
url = 'https://raw.githubusercontent.com/Data-Matt0/marketing-campaign-dashboard/refs/heads/main/data/queries/q4_web_purchases_vs_response.csv'
df = pd.read_csv(url)

# Convert 'response' to categorical labels for clarity
df['response'] = df['response'].map({0: 'No Response', 1: 'Response'})

# Set style and figure size
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Create horizontal barplot with a vibrant color palette
barplot = sns.barplot(
    x='avg_web_purchases',
    y='response',
    data=df,
    palette='mako_r',  # reversed 'mako' for a sleek dark-to-light gradient
    edgecolor='black'
)

# Add value labels inside the bars for readability
for index, row in df.iterrows():
    plt.text(row.avg_web_purchases - 0.1, index, f"{row.avg_web_purchases:.2f}",
             color='white', weight='bold', fontsize=12, ha='right', va='center')

# Add title and axis labels with styling
plt.title('🌐 Average Web Purchases by Campaign Response', fontsize=18, weight='bold', pad=15)
plt.xlabel('Average Number of Web Purchases', fontsize=14)
plt.ylabel('Response Status', fontsize=14)

# Clean up layout
plt.xlim(0, df['avg_web_purchases'].max() * 1.1)  # Add padding to x-axis
plt.tight_layout()

# Show plot


# Save figure as PNG in current directory
plt.savefig('query4_web_purchases_vs_response.png', dpi=300, bbox_inches='tight')

plt.show()
```
